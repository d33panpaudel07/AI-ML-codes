```python
import pandas as pd
customer_call_df = pd.read_excel('../data/Customer Call List.xlsx')
```

```python
import numpy as np
```

```python
# displaying the dataframe
print(customer_call_df)
```

```python
# first dropping the unnecessary column
customer_call_df.drop(columns='Not_Useful_Column',inplace=True)
print(customer_call_df)
```

```python
# starting to fill missing values
# gonna use mean values for average and acceptable data
# customer_call_df.fillna("Temporary Value")

# another approach was to convert all missing data such as N/a or Not found to NaN for easy handling with numpy
customer_call_df = customer_call_df.replace({'N/a':np.nan})
print(customer_call_df)
```

```python
# gonna see the status of data now 
print(customer_call_df)
```

```python
# since the N/a have become a NaN in numpy we can replace those by "Unknown"
customer_call_df['Phone_Number'].fillna("Unknown", inplace=True)
print(customer_call_df)
```

```python
# putting unknown in lastname as well
# customer_call_df['Last_Name'].fillna("Unknown",inplace=True)
customer_call_df['Last_Name'] = customer_call_df['Last_Name'].fillna("Unknown")
# same for address, paying customer and dnc
customer_call_df['Address'] = customer_call_df['Address'].fillna("Unknown")
customer_call_df['Paying Customer'] = customer_call_df['Paying Customer'].fillna("Unknown")
customer_call_df['Do_Not_Contact'] = customer_call_df['Do_Not_Contact'].fillna("Unknown")
customer_call_df['Phone_Number'] = customer_call_df['Phone_Number'].fillna("Unknown")
print(customer_call_df)
```

```python
# now to fix the inconsistency that is in paying custome and do_not_contact
customer_call_df[['Paying Customer','Do_Not_Contact']] = customer_call_df[['Paying Customer','Do_Not_Contact']].replace({'Y':'Yes', 'N':'No'})
print(customer_call_df)
```

```python
# fixing the values
customer_call_df['Last_Name'] = customer_call_df['Last_Name'].str.replace(r'[./]','',regex=True)

customer_call_df['Phone_Number'] = customer_call_df['Phone_Number'].str.replace(r'[-|/]','',regex=True)

print(customer_call_df)
```

```python
# separating phone with -
# customer_call_df['Phone_Number'] = customer_call_df['Phone_Number'].str.extract(r'(\d{3}-\d{3}-\d{4})')
customer_call_df = customer_call_df.replace('Unknown',np.nan)

# making the data frame consistent
# customer_call_df['Phone_Number'] = customer_call_df['Phone_Number'].str.replace(r'[|/]','-',regex=True)
import re

# finally now we can put - 
customer_call_df['Phone_Number'] = customer_call_df['Phone_Number'].apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:]}" if pd.notna(x) and re.fullmatch(r'\d{10}', x) else x)
# customer_call_df['Phone_Number'] = customer_call_df['Phone_Number'].str.extract(r'(\d{3}[-.\s]?\d{3}[-.\s]?\d{4})')



print(customer_call_df)
```

```python
# dropping rows with do_not_contact as yes
customer_call_df = customer_call_df[customer_call_df['Do_Not_Contact'] != 'Yes']
print(customer_call_df)
```

```python
customer_call_df = customer_call_df.dropna()
print(customer_call_df)
```

```python
# fixing column name
customer_call_df = customer_call_df.rename(columns={"Paying Customer": "Paying_Customer"})

# removing duplicates
customer_call_df = customer_call_df.drop_duplicates()
print(customer_call_df)
```

```python
# now to finally export the cleaned data
customer_call_df.to_excel('../data/cleaned_customer_call.xlsx',index=False)
```
