## Group by and Aggregating

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/reban87/ML-From-Scratch/refs/heads/main/introduction_to_ML/data_cleaning/Flavors.csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor')
print(group_by_frame)
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
group_by_frame.sum()
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'max', 'count']})
```

```python
group_by_frame.describe()
```

## Merge, Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/reban87/ML-From-Scratch/refs/heads/main/introduction_to_ML/data_cleaning/LOTR.csv')
df1
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/reban87/ML-From-Scratch/refs/heads/main/introduction_to_ML/data_cleaning/LOTR%202.csv')
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1, df2],ignore_index=True)
```

```python
# Iner join
df1.merge(df2, how= 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
# Outer Join
df1.merge(df2, how= 'outer')
```

```python
df1.merge(df2, how= 'right')
```

```python
df1.merge(df2, how= 'left')
```
