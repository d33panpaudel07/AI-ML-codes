```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basics of matplotlib
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.linspace(0,10,100)
x2 = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x2)
plt.plot(x,y,label = "Sine wave")
plt.plot(x2,y2,label = "Cos wave")
plt.title("Line plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Bar chart
categories = ["A", "B", "C"]
values = [10, 20, 30]
plt.bar(categories, values, color = ["blue", "maroon", "green"])
# plt.bar(categories, values, color = ["blue", "maroon", "green"], edgecolor = "pink", linewidth = 25)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show
```

# Visualization with seaborn

```python
import seaborn as sns

# HIstogram and KDE plot
data = np.random.randn(1000)
sns.histplot(data, kde = True, color = "green")
plt.title("Histogram with KDE")
plt.show
```

```python
# Heatmap
import random


random_integers = [[random.randint(1, 10) for _ in range(5)] for _ in range(5)] 
print("Random Matrix:\n", random_integers)
# matrix = np.random.rand(5,5)
# print("Random Matrix:\n", matrix)
```

```python
sns.heatmap(random_integers, annot=True, cmap = "YlOrBr")
plt.title("Heatmap Example")
plt.show()
```

```python
# pairplot example
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.show
```
